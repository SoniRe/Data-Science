```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
vocab_size = 10000

tokenizer = Tokenizer(num_words=vocab_size)
tokenizer.fit_on_texts(quotes)
```

```python
word_index = tokenizer.word_index
print("Length : ", len(word_index))
list(word_index.items())[:10]
```

```python
sequence = tokenizer.texts_to_sequences(quotes)
```

```python
quotes[0]
```

```python
sequence[0]
```

```python
X = []
y = []

for seq in sequence:
    for i in range(1, len(seq)):
        input_seq = seq[:i]
        output_seq = seq[i]
        X.append(input_seq)
        y.append(output_seq)
```

```python
X[:20]
```

```python
y
```

```python
from tensorflow.keras.utils import pad_sequences

max_len = max(len(x) for x in X)

X_padded = pad_sequences(X, maxlen=max_len, padding='pre')
X_padded
```

```python
y = np.array(y)
y
```

```python
from tensorflow.keras.utils import to_categorical
y_one_hot = to_categorical(y, num_classes = vocab_size)
```

```python
y_one_hot.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, SimpleRNN
import warnings
warnings.filterwarnings('ignore')
```

```python
embeddings_dim = 100
rnn_units = 128
```

```python
rnn_model = Sequential()

rnn_model.add(
    Embedding(input_dim=vocab_size, output_dim=embeddings_dim,
             input_length=max_len)
)

rnn_model.add(SimpleRNN(units=rnn_units))

rnn_model.add(Dense(units=vocab_size, activation='softmax'))
```

```python
rnn_model.compile(
    optimizer='adam', 
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
rnn_model.summary()
```

```python
# LSTM
lstm_model = Sequential()

lstm_model.add(
     Embedding(input_dim=vocab_size, output_dim=embeddings_dim,
             input_length=max_len)
)

lstm_model.add(
    LSTM(units=rnn_units)
)

lstm_model.add(Dense(units=vocab_size, activation='softmax'))
```

```python
lstm_model.compile(
    optimizer='adam', 
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Using Saved LSTM Model
index_to_word = {}
for word, index in word_index.items():
    index_to_word[index] = word
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences

lstm_model_load = load_model('lstm_model.h5')
```

```python
# Single Word Predictor

def predictor(model, tokenizer, text, max_len):
    text = text.lower()
    
    seq = tokenizer.texts_to_sequences([text])[0]
    seq = pad_sequences([seq], maxlen=max_len, padding='pre')

    pred = model.predict(seq, verbose=0)
    pred_index = np.argmax(pred)
    return index_to_word[pred_index]
```

```python
seed_text = "you will"
next_word = predictor(lstm_model_load, tokenizer,seed_text ,max_len)

print(next_word)
```

```python
# Generate Text

def generate_text(model, tokenizer, text, max_len, n_words):
    for _ in range(n_words):
        next_word = predictor(model, tokenizer, text, max_len)
        if next_word == "":
            break
        text += " " + next_word
        return text
```

```python
seed = "the meaning of life"
generate_text = generate_text(lstm_model_load, tokenizer, seed, max_len, 10)
print(generate_text)
```
