```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
#s -> size of the bubble, alpha = 0.5 -> semi-transparent, 0.2 very light
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*10, alpha=0.5)
```

```python
#Pie Chart
label = ['A', 'B', 'C', 'D']
values = [10, 20, 40, 30]
plt.pie(values, labels=label, wedgeprops={'width':1})
```

```python

#If use width:0.4 then doughnut
plt.pie(values, labels=label, wedgeprops={'width':.5})
```

```python
data = {
    "Country": ["IND", "USA", "CHN", "BRZ"],
    "Values": [121221,2323,23123,12313]
}

df = pd.DataFrame(data)

fig = px.choropleth(df, locations="Country", color='Values')
fig.show()
```
