```python
import seaborn as sns
import matplotlib.pyplot as plt
```

We use distribution plots to visualize the distribution of quantitative data.

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

distplot/histplot
jointplot
pairplot
rugplot

```python
df = sns.load_dataset('tips') 
df
```

```python
# Distribution Representation
# sns.displot(df['total_bill'],bins=20)
# Same as histplot
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins=20, kde=True)

plt.subplot(1,2,2)
sns.histplot(df['tip'], bins=30, kde=True)
# Kernel Density Estimator -> Tells data skewness
```

```python
# Joint plot to compare
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')
```

```python
# Pair Plot -> Pairs multiple value
sns.pairplot(df, hue='size',palette='rainbow')
```

```python
# Rug Plot
sns.rugplot(df['total_bill'])
```
