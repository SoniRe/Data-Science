**Object Oriented Way**

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(20,8), dpi=100)
# Left and Bottom
axis1 = fig.add_axes([0,0,1,1])
axis1.plot(x_axis, y_axis)
axis2 = fig.add_axes([0.2,0.2,0.5,0.5])
axis2.plot(x_axis, y_axis)
```

**Types of Plot**

```python
# Saving Plot
fig = plt.figure()
axis3 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis3.plot(x_axis,y_axis)
plt.savefig("BasicPlot.png")
```
