```python
 import numpy as np
```

```python
#Numpy Indexing and Slicing of Vectors
arr = np.arange(11,21)
arr
```

```python
arr[1]
```

```python
arr_slice = arr[3::2]
arr_slice
```

Numpy Indexing and Slicing of Matrix

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0]
```

```python
arr[0][3]
```

```python
arr[0:2,1:3] 
```

```python
arr[3:,3:]
```

```python
arr[:,2]
```
