```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, Holt
```

```python
df = pd.read_csv('AirPassengers.csv')
df.head()
```

```python
df.columns = ['Date', 'Number of Passengers']
df.head()
```

```python
plt.plot(df['Date'], df['Number of Passengers'])
```

```python
#ADF Test
def checkStationary(series):
    result = adfuller(series)
    print("ADF Statistics : ", result[0])
    print("P-value : ", result[1])   

    if(result[1] < 0.05):
        print("Stationary")
    else:
        print("Non Stationary")
```

```python
checkStationary(df['Number of Passengers'])
```

```python
df["Diff"] = df["Number of Passengers"].diff()
df["Diff"].head()
```

```python
df.dropna(inplace=True)
df['Diff'].head()
```

```python
checkStationary(df['Diff'])
```

```python
df['DoubleDiff'] = df['Diff'].diff()
df['DoubleDiff'].head()
```

```python
df.dropna(inplace=True)
df['DoubleDiff'].head()
```

```python
checkStationary(df['DoubleDiff'])
```

```python
plt.plot(df['DoubleDiff'])
```

```python
#Auto Regression - Predict NEXT Value using Past values
model_ar = AutoReg(df['DoubleDiff'], lags=12)

result = model_ar.fit()

#Lower AIC and BIC the better
print("AIC : ", result.aic)
print("BIC : ", result.bic)
```

```python
#Moving Average - Predict next value using Past FORECAST errors
# p = AR Terms, d = differencing, q = MA
# Smooths our random noice p = 0, d = 0, q = 1
model_ma = ARIMA(df['DoubleDiff'], order=(0, 0, 1))

result = model_ma.fit()
print("AIC : ", result.aic)
print("BIC : ", result.bic)
```

```python
#Exponential Smoothing
#Weighted Avg - Recent value matter more
#Smoothening Level => 0 = Trust Old Data, 1 = Trust only last data

#Single Exponential Smoothening
model_ses = SimpleExpSmoothing(df['DoubleDiff'])

result = model_ses.fit(smoothing_level=0.2, optimized=False)
print("AIC : ", result.aic)
print("BIC : ", result.bic)
```

```python
#Double Exponential Smoothening
#Handles data with Trend
model_des = Holt(df['Number of Passengers'])

result = model_des.fit(smoothing_level=0.2, smoothing_trend=0.1, optimized=False)
print("AIC : ", result.aic)
print("BIC : ", result.bic)
```

```python
#ARIMA
model_arma = ARIMA(df['DoubleDiff'], order=(1,0,1))

result = model_arma.fit()
print("AIC : ", result.aic)
print("BIC : ", result.bic)
```

```python
#ARIMA
model_arima = ARIMA(df['Number of Passengers'], order=(1,1,1))

result = model_arima.fit()
print("AIC : ", result.aic)
print("BIC : ", result.bic)

forecast = result.forecast(steps=12)
plt.figure(figsize=(10,4))
plt.plot(df['Number of Passengers'])
plt.plot(forecast)
```
