```python
Find Missing Data
```

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}

df = pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
Remove Missing Data [Works on Row Basis]
```

```python
df.dropna()
```

```python
df
```

```python
#Should have atleast 3 non null values
df.dropna(thresh=3)
```

```python
Filling the Missing Data
```

```python
df.fillna(0)
```

```python
#Fill with Specific
values = {'A':0,'B':100,'C':300,'D':400}
df.fillna(value=values)
```

```python
#Fill with Mean Values
df.fillna(df.mean())
```
