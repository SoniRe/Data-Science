```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
df
```

 **Group By**

```python
#Group By Category and calculate sum of Sales
cat = df.groupby('Category')['Sales'].sum()
cat
```

```python
#Group by Store and Calculate the sum of Sales
cat2 = df.groupby('Store')['Sales'].sum()
cat2
```

```python
#Qualitative - Categorical Data (Group By)
# Quantative - Numberical Data
#Group by multiple columns
cat = df.groupby(['Category','Store'])['Sales'].sum()
cat
```

**Aggregation**

```python
df['Sales'].mean()
```

```python
df['Sales'].median()
```

```python
df['Sales'].mode()
```

```python
# Standard Daviation
df['Sales'].std()
```

```python
df['Sales'].agg(['mean','min','max','std','sum','count'])
```
